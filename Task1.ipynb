```python
#Task 1.2 Create a random 100000x1 array and visualize
import matplotlib.pyplot as plt
```

```python
plt.hist(np.random.rand(100000), density=True, bins=100, histtype="step", color="blue", label="rand")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc = "upper left")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
#Experiment by changing bins value
plt.hist(np.random.rand(100000), density=True, bins=1000, histtype="step", color="blue", label="rand")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc = "upper left")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
#Task 1.3 Z = X**2 + Y**2 in a 3D Plot

from mpl_toolkits.mplot3d import Axes3D
```

```python
X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('Surface plot of Z = X^2 + Y^2')

plt.show()
```
