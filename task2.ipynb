```python
#Task 2

import pandas as pd

#load the data
df = pd.read_csv('/content/Au_nanoparticle_dataset.csv')
```

```python
#2.1 New filtered dataframe
new_df = df.drop(columns=['N_total', 'N_bulk', 'N_surface', 'R_avg'])
```

```python
#2.2 First 20 samples
new_df.head(20)
```

```python
#Task 2.3 Statistics

mean_values = new_df.mean()
std_values = new_df.std()
quartiles = new_df.quantile([0.25, 0.5, 0.75])

print("Mean values:\n", mean_values)
print("\nStandard Deviation values:\n", std_values)
print("\nQuartile values:\n", quartiles)
```

```python
#Task 3.4 Plot histogram
import matplotlib.pyplot as plt
```

```python

fig, axs = plt.subplots(1, 4, figsize=(20, 5))

# Use columns that exist in the new_df DataFrame
features = new_df.columns[:4]  # Assuming you want to plot the first 4 columns

for i, feature in enumerate(features):
    axs[i].hist(new_df[feature], bins=20, alpha=0.7)
    axs[i].set_title(f'Histogram of {feature}')
    axs[i].set_xlabel(feature)
    axs[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Task 2.5 Plot histograms in a 1x4 layout
import seaborn as sns

g = sns.PairGrid(new_df)

g.map_upper(sns.histplot)
g.map_diag(sns.histplot, kde=True)
g.map_lower(sns.kdeplot)

plt.show()
```
